### Problem 22. The Yukawa potential.

#### The Yukawa potential is given by $$V(r)=-k\frac{e^{-\alpha r}}{r},$$ where $k$ and $\alpha$ are constants.

### Problem 23. Other potentials.

#### A particle of mass $m$ moves under the influence of hte central potential $$V(r)=-\frac{k}{r^4},$$ where $k$ is a constant and $k>0$.  At time $t=0$ the particle is at $r=r_0$ and is moving with a velocity $v_0$ directed at an angle of $45^{\circ}$ with respect to the radial outward direction.

#### (b) Make a plot of the effective potential for the radial motion of the particle.  Does the potential support circular orbits?  If so, are these orbits stable?

```python
'''Required Libraries'''
import matplotlib.pyplot as plt
import numpy as np
```

#### 22 (a) Write the Lagrangian for this central potential and find the equation of motion as well as the effective potential.

```python
'''Constants'''
l = 1   #angular momentum [J * s]
m = 1   #mass [kg]
k = 1   #scaling constant [N * m**2]
r0 = 1  #initial radius [m]

'''Decay terms'''
alpha1 = 0
alpha2 = 1
alpha3 = 10

'''Radius interval'''
r_points = np.linspace(0.1, 2 * r0, 100, endpoint = True)

'''Function to calculate the Yukawa potential'''
def Yukawa(alpha, r):
    V = -k * np.exp(-alpha * r) / r
    return V

'''Calculate Yukawa and effective potentials'''
yukawa1 = [Yukawa(alpha1, r) for r in r_points]
yukawa2 = [Yukawa(alpha2, r) for r in r_points]
yukawa3 = [Yukawa(alpha3, r) for r in r_points]
Veff1 = [((l**2) / (2 * m**2 * r**2)) + Yukawa(alpha1, r) for r in r_points]
Veff2 = [((l**2) / (2 * m**2 * r**2)) + Yukawa(alpha2, r) for r in r_points]
Veff3 = [((l**2) / (2 * m**2 * r**2)) + Yukawa(alpha3, r) for r in r_points]

'''Create figure and plot results'''
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (14, 5))
fig.tight_layout(pad  =  2.0)

ax1.set_xlabel('Radius $r\ /\ r_0$')
ax1.set_ylabel('Potential Energy [J]')
ax1.set_title('Yukawa Potential $(k = r_0 = 1)$')
ax1.grid(True, linestyle = 'dashed', color = 'darkgray')
ax1.plot(r_points, yukawa1, linestyle = 'solid', color = 'limegreen', label = f'\u03B1 = {alpha1}')
ax1.plot(r_points, yukawa2, linestyle = 'solid', color = 'dodgerblue', label = f'\u03B1 = {alpha2}')
ax1.plot(r_points, yukawa3, linestyle = 'solid', color = 'mediumorchid', label = f'\u03B1 = {alpha3}')
ax1.legend(prop = {'size': 12}, loc = 'lower right', ncol = 1)

ax2.set_xlabel('Radius $r\ /\ r_0$')
ax2.set_ylabel('Potential Energy [J]')
ax2.set_title('Effective Yukawa Potential $(l = m = k = r_0 = 1)$')
ax2.grid(True, linestyle = 'dashed', color = 'darkgray')
ax2.plot(r_points, Veff1, linestyle = 'solid', color = 'limegreen', label = f'\u03B1 = {alpha1}')
ax2.plot(r_points, Veff2, linestyle = 'solid', color = 'dodgerblue', label = f'\u03B1 = {alpha2}')
ax2.plot(r_points, Veff3, linestyle = 'solid', color = 'mediumorchid', label = f'\u03B1 = {alpha3}')
ax2.legend(prop = {'size': 12}, loc = 'upper right', ncol = 1)
fig.savefig('c:/Users/Tucker Knaak/Downloads/PHY611_HW5a22_Plot.png', bbox_inches = 'tight')
```

#### 23 (b) Make a plot of the effective potential for the radial motion of the particle.  Does the potential support circular orbits?  If so, are these orbits stable?

```python
'''Constants'''
l = 1   #angular momentum [J * s]
m = 1   #mass [kg]
r0 = 1  #initial radius [m]

'''Scaling constant terms'''
k1 = 0.01
k2 = 0.001
k3 = 0.0001

'''Radius interval'''
r_points = np.linspace(0.1, 2 * r0, 100, endpoint = True)

'''Function to calculate the central potential'''
def central(k, r):
    V = -k / (r**4)
    return V

'''Calculate central and effective potentials'''
central1 = [central(k1, r) for r in r_points]
central2 = [central(k2, r) for r in r_points]
central3 = [central(k3, r) for r in r_points]
Veff1 = [((l**2) / (2 * m**2 * r**2)) + central(k1, r) for r in r_points]
Veff2 = [((l**2) / (2 * m**2 * r**2)) + central(k2, r) for r in r_points]
Veff3 = [((l**2) / (2 * m**2 * r**2)) + central(k3, r) for r in r_points]

'''Create figure and plot results'''
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (14, 5))
fig.tight_layout(pad  =  2.0)

ax1.set_xlabel('Radius $r\ /\ r_0$')
ax1.set_ylabel('Potential Energy [J]')
ax1.set_title('Central Potential $(r_0 = 1)$')
ax1.grid(True, linestyle = 'dashed', color = 'darkgray')
ax1.plot(r_points, central1, linestyle = 'solid', color = 'limegreen', label = f'k = {k1}')
ax1.plot(r_points, central2, linestyle = 'solid', color = 'dodgerblue', label = f'k = {k2}')
ax1.plot(r_points, central3, linestyle = 'solid', color = 'mediumorchid', label = f'k = {k3}')
ax1.legend(prop = {'size': 12}, loc = 'lower right', ncol = 1)

ax2.set_xlabel('Radius $r\ /\ r_0$')
ax2.set_ylabel('Potential Energy [J]')
ax2.set_title('Effective Central Potential $(l = m = r_0 = 1)$')
ax2.grid(True, linestyle = 'dashed', color = 'darkgray')
ax2.plot(r_points, Veff1, linestyle = 'solid', color = 'limegreen', label = f'k = {k1}')
ax2.plot(r_points, Veff2, linestyle = 'solid', color = 'dodgerblue', label = f'k = {k2}')
ax2.plot(r_points, Veff3, linestyle = 'solid', color = 'mediumorchid', label = f'k = {k3}')
ax2.legend(prop = {'size': 12}, loc = 'upper right', ncol = 1)
fig.savefig('c:/Users/Tucker Knaak/Downloads/PHY611_HW5a23_Plot.png', bbox_inches = 'tight')
```
